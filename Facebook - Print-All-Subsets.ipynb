```python
def capitalize(string):
    return string.upper()


def binary_num(n, padding):
    unpadded_bin = str(bin(n))[2:]

    padding_length = max(0, padding - len(unpadded_bin))

    return padding_length*'0' + unpadded_bin


def print_subsets(S):

    # Get the length of S
    n = len(S)

    for i in range(2**n):
        binary_number = binary_num(i, n)

        subset = ""
        for index, letter in enumerate(binary_number):
            if letter == "1":
                subset += str(S[index]) + ","
        print(subset)
```
